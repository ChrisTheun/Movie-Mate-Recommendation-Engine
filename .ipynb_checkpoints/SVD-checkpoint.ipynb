```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.sparse
import seaborn as sns
%matplotlib inline
```

```python
movies = pd.read_csv("movies.csv") # details on the movies in the movielens dataset
ratings = pd.read_csv("ratings.csv") #the ratings the movielens community gave to movies
chris = pd.read_csv("chris ratings.csv") # my own ratings exported from movielens
ratings_table = pd.read_csv("ratings_table.csv", sep = "\t", index_col = 0)
coocc = pd.read_csv("coocc.csv", sep = "\t", index_col = 0)
```

```python
# Convert the rating column from float64 to float32 to save space and convert movieId and userId from float64 to objects (str)
ratings["rating"] = ratings["rating"].astype("float32")
ratings["userId"] = ratings["userId"].astype("object")
ratings["movieId"] = ratings["movieId"].astype("object")
movies["movieId"] = movies["movieId"].astype("object")
chris["movie_id"] = chris["movie_id"].astype("object")

chris["userId"] = "Chris" # Add userId to identify my own ratings when I add them to the ratings df later
chris = chris[["userId", "movie_id", "rating"]] # Only keep the userId, movie_id and rating columns
chris.columns = ['userId', 'movieId', 'rating'] # Rename movie_id to movieId so it is consistent with the ratings df

coocc.index = list(ratings_table.index)
coocc.columns = list(ratings_table.index)

# Remove movies in rated that are not rated by anyone else (e.g. because they weren't yet released)
chris = chris[chris["movieId"].isin(list(ratings["movieId"]))]

# Drop the timestamp column in the ratings dataframe, because we do not use it
ratings = ratings.drop("timestamp", axis = 1)

# Merge movie information (especially title) with the ratings for each movie
movieratings = pd.merge(ratings, movies, on = "movieId")
```

#### Normalize the user ratings

We center the ratings by subtracting the user's mean rating from every single one of his/her individiual ratings.
This demeaning process takes into account the users that are disproportionally lenient or harsh with their ratings.

```python
ratings_mat = ratings_table.as_matrix()
ratings_mean = np.nanmean(ratings_mat,axis=1)
norm_ratings = ratings_mat - ratings_mean.reshape(-1,1)
scaled_ratings = pd.DataFrame(norm_ratings).fillna(0)
scaled_ratings.index = ratings_table.index
scaled_ratings.columns = ratings_table.columns
```

## Singular Value Decomposition (matrix factorization)

SVD lets us break down the user-item matrix (A) into three matrices (U, S and V) that can be combined to approximate the original user-item matrix according to the following equation:

A = USV^T

A is an m × n matrix
U is an m × n orthogonal matrix
S is an n × n diagonal matrix
V is an n × n orthogonal matrix

n (denoted as k in the svds function) can be seen as the number of features that represent the dimensions on which movies can vary. For example, if we say n = 50, then we will allow for 50 dimensions which each comprise an important element of the movies. One dimension could be an indication of how action-packed a movie is, while another dimension could indicate whether it is a big or small budget movie. The interesting thing is that SVD computes these dimensions and therefore they do not have to be specified by the researcher or users. The side-effect is that the meaning of the dimensions can be hard to interpret. V is in this case regarded as the matrix which stores the values for these dimensions for each movie. The U matrix then gives us the degree to which user values each of these dimension. Consequently, the U matrix can be seen as the taste profile for each user. 

We use this U matrix to compute the differences in taste between the users and try to find which users are closest to the selected user (myself in this case). We do this by using the absolute, euclidean and cosine distance metrics. The user with the lowest distance can be regarded as the one that matches me the best in terms of movie taste.

```python
from scipy.sparse.linalg import svds
from scipy.spatial.distance import cdist
```

Define function which outputs the 10 most similar users based on SVD

```python
def SimilarUsersSVD(user_item_mat, n_features, coocc_mat, dist_metric, main_user_id, min_cooccurrences = 100 , n_most_similar_users = 10):
    
    U, s, V = svds(user_item_mat, k = n_features) # Decompose the user-item matrix into 3 matrices that approximate the original using k dimensions
    
    print("U shape:", U.shape) # user taste profiles
    print("s shape:", s.shape)
    print("V shape:", V.shape) # movie features
    
    U_df = pd.DataFrame(U) # cast the U matrix with the taste profiles into a dataframe
    U_df.index = list(user_item_mat.index) # make sure the indices correspond to those of the users in the user-item matrix
    
    
    # Compute distance matrix by using either the absolute, eucildean or cosine distance metric
    if dist_metric == "absolute":
        dist_mat = cdist(U_df, U_df, lambda u, v: np.nansum(abs((u-v))))
    elif dist_metric == "euclidean":
        dist_mat = cdist(U_df, U_df, lambda u, v: np.sqrt(np.nansum((u-v)**2)))
    elif dist_metric == "cosine":
        dist_mat = cdist(U_df.fillna(0), U_df.fillna(0), "cosine")

    # Cast the results into a dataframe
    dist = pd.DataFrame(dist_mat)
    
    # Make sure both the row index and the column names represent the user IDs
    dist.index = list(user_item_mat.index)
    dist.columns = list(user_item_mat.index)

    main_distances = list(dist.loc[str(main_user_id)]) # Cast the distances between the main user and the other users in a list
    ratings_shared = list(coocc.loc[str(main_user_id)]) # Cast the counts of movies both rated by the main user and the other user (cooccurrences) in a list

    # Create a dataframe with these two lists as columns
    main_dist = pd.DataFrame(data = [main_distances, ratings_shared]).transpose()
    main_dist.index = user_item_mat.index
    main_dist.columns = ["dist", "cooccurrences"]
    main_dist = main_dist.drop(str(main_user_id)) # Drop the distance of the main user to itself

    # Filter out the users that have not seen enough of the same movies as the main user
    main_dist = main_dist[main_dist["cooccurrences"] >= min_cooccurrences] 

    # Return the 10 users with the smallest distance to the main user
    return(main_dist.nsmallest(n_most_similar_users, "dist"))
    
    
```

Apply the defined function and compare the selected user's (my own) ratings to those of the user which has the lowest distance to the selected user in order to see if their movie taste seems to be similar

#### Closest absolute distance (k = 10)

```python
closest_abs_dist = SimilarUsersSVD(user_item_mat = scaled_ratings, n_features = 10, coocc_mat = coocc, dist_metric = "absolute", main_user_id = "Chris")
print(closest_abs_dist)
plt.hist(ratings["rating"][ratings["userId"] == closest_abs_dist.index[0]], bins = 9, edgecolor='black', linewidth=1.2)
# Inspect the ratings of the user closest to me to see if his/her taste matches with my own
pd.merge(movieratings[movieratings["userId"] == closest_abs_dist.index[0]].sort_values("rating", ascending = False), chris, on = "movieId")
```
